En géométrie, un triangle est constructible si la somme de deux côtés est strictement supérieure au troisième côté.

Écrivez une fonction `est_triangle(a, b, c)` qui renvoie `True` si les trois longueurs permettent de former un triangle, et `False` sinon.

```python
def est_triangle(a, b, c):
    ''' Renvoie True si les longueurs a, b et c peuvent former un triangle, False sinon '''
    if a + b > c and a + c > b and b + c > a:
        return True
    else:
        return False
    
assert est_triangle(3, 4, 5) == True
assert est_triangle(1, 2, 3) == False
assert est_triangle(5, 5, 5) == True
assert est_triangle(6, 8, 4) == True
assert est_triangle(2, 3, 4) == True
assert est_triangle(6, 2, 3) == False
assert est_triangle(1, 5, 3) == False
```
