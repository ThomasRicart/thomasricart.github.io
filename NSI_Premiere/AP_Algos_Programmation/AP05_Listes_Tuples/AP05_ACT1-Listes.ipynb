### **Introduction : Les listes**

    
Une liste est une structure de donnée linéaire qui contient des `éléments` ou `items` de type quelconque. (int, str, tuple, list).
    
Ces éléments sont placés à l'intérieur de crochets et séparés par les des virgules.
    
On peut représenter une liste comme un tableau de ce type. Ici la liste possède 8 éléments d'indice compris entre 0 et 7
    
| Indice i  | 0 | 1 | 2 | 3 | 4 | 5 | 6 | 7 |
| :-------: |:-:|:-:|:-:|:-:|:-:|:-:|:-:|:-:|
| Elément   | 4 | 6 | 3 | 1 | 2 | 7 | 8 | 9 |     
    
    
En python, une liste est de type `list`
    
Les listes sont des variables **mutables**. On  peut donc les modifier une fois qu'elles ont été créées


```python
L = [2, 4, 5]
print(L[0]) # Affiche 2
print(L[1]) # Affiche 4
print(L[2]) # Affiche 5
print(type(L)) # Affiche <class 'list'>
print(len(L)) # Affiche 3
```
