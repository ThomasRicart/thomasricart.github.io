## 🏆 Défi Expert : Statistiques de liste
Crée une fonction `analyser_liste(nombres)` qui prend une liste de nombres en entrée et retourne un **dictionnaire** contenant :
- Le nombre d'éléments.
- La valeur maximale.
- La moyenne.

*(Indice : utilise les fonctions intégrées `len()`, `max()`, et `sum()`)*

````python
def analyser_liste(nombres):
    stats = {
        "quantité": len(nombres),
        "maximum": max(nombres),
        "moyenne": sum(nombres) / len(nombres)
    }
    return stats

# Test
ma_liste = [10, 20, 30, 40, 50]
print(analyser_liste(ma_liste))
````
