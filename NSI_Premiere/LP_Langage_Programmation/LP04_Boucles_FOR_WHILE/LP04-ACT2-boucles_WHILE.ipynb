### **Introduction : La boucle WHILE**

La boucle WHILE est une **boucle non bornée**. On ne sait avant de commencer combien de boucle nous allons effectuer.

```python
while condition:
    instruction1
instruction2
```


`condition` est un booléen qui est évalué à chaque boucle.
* si `condition` est évalué à `True`, on effectue la boucle et `instruction1`
* si `condition` est évalué à `False`, on n'effectue pas la boucle et on effectue `instruction2`.
