2. Maintenant que nous avons la liste de dictionnaire nous allons créer un fichier CSV de fusion. Vérifier que la fusion a été correcte.


````python
def make_CSV(table_pilotes):
    ''' Cette fonction prend en entrée une table de pilotes (liste de dictionnaires) et crée un fichier CSV `pilotes.csv` contenant les données de cette table.'''
    with open('pilotes.csv', 'w', newline='', encoding='utf-8') as f:
        writer = csv.DictWriter(f, fieldnames=table_pilotes[0].keys(), delimiter=';')
        writer.writeheader()
        writer.writerows(table_pilotes)

make_CSV(table_pilotes)
````
