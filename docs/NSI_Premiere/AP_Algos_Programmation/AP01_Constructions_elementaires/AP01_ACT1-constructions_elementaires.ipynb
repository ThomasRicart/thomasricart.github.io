## 0. Quelques précisions

En python, il est possible d'afficher:
* du texte (chaines de caractères)
* des valeurs de variables

```python
>>> print('Bonjour')

>>> mot = 'Hello'
>>> print(mot)
Hello
>>> valeur = 6
>>> print(valeur)
6
```
