### **Introduction : Présentation**

Un objet est un tuple: (nom, poids, prix), par exemple:

```python
epee = ("Épée en fer", 10, 50)
potion = ("Potion de soin", 1, 15)
```

Un inventaire est une liste d'objets

```python
inventaire = [epee, potion]
```

**Exercice 1 :**

Ecrire une fonction `calcule_poids(inventaire)` qui renvoie le poids total de l'inventaire.


````python
POIDS_max = 30
epee = ("Épée en fer", 10, 50)
potion = ("Potion de soin", 1, 15)
inventaire = [epee, potion]

def calcule_poids(inventaire):
    """Retourne le poids total des objets de l'inventaire."""
    # A compléter
    total = 0
    for objet in inventaire:
        total += objet[1]
    return total


assert calcule_poids(inventaire) == 11
````

**Exercice 2 :** 

Ecrire une fonction `calcule_prix(inventaire)` qui renvoie le prix total de l'`inventaire`



````python
def calcule_prix(inventaire):
    """Retourne le prix total des objets de l'inventaire."""
    # A compléter
    total = 0
    for objet in inventaire:
        total += objet[2]
    return total
assert calcule_prix(inventaire) == 65
````

**Exercice 3 :**
Ecrire une fonction `ramasser_objet(inventaire, objet)` qui ajoute `objet` à `inventaire` si le poids total est inférieur à un `poids_max`



````python
def ramasser_objet(inventaire, objet):
    """Ajoute un objet à l'inventaire."""
    # A compléter
    if calcule_poids(inventaire) + objet[1] <= POIDS_max:
        inventaire.append(objet)
        



casque = ("Casque en cuir", 5, 20)
ramasser_objet(inventaire, casque)
# L'objet est ajouté à l'inventaire car le poids total (16) ne dépasse pas le poids maximum (30)
assert calcule_poids(inventaire) == 16
assert calcule_prix(inventaire) == 85

armure = ("Armure en acier", 25, 100)
ramasser_objet(inventaire, armure)
# L'objet n'est pas ajouté à l'inventaire car le poids total (36) dépasse le poids maximum (30)
assert calcule_poids(inventaire) == 16
assert calcule_prix(inventaire) == 85
````

**Exercice 4 :**

Ecrire une fonction `jeter_objet(inventaire, objet)` qui supprime `objet` de l'`inventaire` s'il existe. Ne fais rien sinon.


````python
def jeter_objet(inventaire, nom_objet):
    """Supprime un objet de l'inventaire par son nom."""
    # A compléter
    for objet in inventaire:
        if objet[0] == nom_objet:
            inventaire.remove(objet)
            break   

    

jeter_objet(inventaire, "Potion de soin")
# L'objet est supprimé de l'inventaire
assert calcule_poids(inventaire) == 15
assert calcule_prix(inventaire) == 70
jeter_objet(inventaire, "Épée en or")
# L'objet n'est pas trouvé dans l'inventaire, donc rien n'est supprimé
assert calcule_poids(inventaire) == 15
assert calcule_prix(inventaire) == 70
````

**Exercice 4 :**

On crée un personne du type
`(nom, inventaire, argent_possede)`

Par exemple:

```python
Alice = ['Alice', [epee, potion], 35]
```

Ici le personnage s'appelle `'Alice'`, possède un inventaire composé d'une épée et d'une potion (definies plus haut) et possède 35 pièces d'or

Créer une fonction `achete_objet(personnage, objet)`

Si le personnage a assez d'argent et que le poids max n'est pas atteint, alors l'objet est ajouté à son inventaire et son nombre de pièces d'or est mis à jour.

````python
Alice = ['Alice', [epee, potion], 35]

def achete_objet(joueur, objet):
    """Ajoute un objet à l'inventaire du joueur si le poids total ne dépasse pas POIDS_max et que le joueur a assez d'argent."""
    # A compléter
    if calcule_poids(joueur[1]) + objet[1] <= POIDS_max and joueur[2] >= objet[2]:
        joueur[1].append(objet)
        joueur[2] = joueur[2] - objet[2]
    







armure = ("Armure en acier", 25, 100)
achete_objet(Alice, armure)
# L'objet n'est pas ajouté à l'inventaire d'Alice car le poids total (36) dépasse le poids maximum (30)
assert calcule_poids(Alice[1]) == 11
assert calcule_prix(Alice[1]) == 65

casque = ("Casque en cuir", 5, 20)
achete_objet(Alice, casque)
# L'objet est ajouté à l'inventaire d'Alice car le poids total (16) ne dépasse pas le poids maximum (30) et Alice a assez d'argent (35 >= 85)
assert calcule_poids(Alice[1]) == 16
assert calcule_prix(Alice[1]) == 85
assert Alice[2] == 15 # Alice a dépensé 20 pièces d'or pour acheter le casque, il lui reste donc 15 pièces d'or


print(Alice[1]) # Affiche l'inventaire d'Alice
````

**Exercice 5 :**

Une équipe de personnages est une liste de personnages. Par exemple

```python
equipe = [Alice, Bob, Charles, Dana]
```

Ecrire une fonction `plus_riche(equipe)` qui renvoie le nom du personnage le plus riche (inventaire et argent possédé)

````python
Alice = ('Alice', [epee, potion], 35)
Bob = ('Bob', [casque, armure], 20)
Charles = ('Charles', [armure, potion, potion], 50)
Dina = ('Dina', [], 10)
equipe = [Alice, Bob, Charles, Dina]

def plus_riche(equipe):
    """Retourne le nom du joueur le plus riche de l'équipe."""
    # A compléter
    plus_riche = equipe[0][0]
    for joueur in equipe:
        if joueur[2] + > plus_riche[2]:
            plus_riche = joueur
        return plus_riche[0]    


assert plus_riche(equipe) == 'Charles'
````
