# ⚔️ Projet NSI : L'Arène des Guerriers
**Objectif :** Développer un simulateur de combat en utilisant des dictionnaires, des listes et vos algorithmes de tri.

### 📜 Les caractéristiques
Chaque guerrier possède :
* **PV** : Points de vie.
* **Force** : Puissance d'attaque.
* **Agilité** : Capacité à esquiver (entre 1 et 50).
* **Armure** : Réduction des dégâts reçus.

Voici un exemple de guerriers pré-tirés pour les tests

**Question 1 :**

Créer la fonction `calculer_score(guerrier)` qui renvoie la somme des caractéristiques du `guerrier`

```python
def calculer_score(guerrier):
    # Somme des caractéristiques PV + Force + Agilité + Armure 
    return guerrier['pv'] + guerrier['force'] + guerrier['agilite'] + guerrier['armure']

# Test avec Aethelgard
Aethelgard = equipe[0]
print(calculer_score(Aethelgard)) # Doit afficher 150
```

**Question 2 :**

Créer la fonction `creer_guerrier(equipe)` qui demande à l'utilisateur de saisir les caractéristiques d'un guerrier et le rajoute à l'équipe. 

```python
def creer_guerrier(equipe):
    """Demande les saisies à l'utilisateur et ajoute le dictionnaire à l'équipe."""
    nom = input("Entrez le nom du guerrier : ")
    pv = int(input("Points de vie (PV) : "))
    force = int(input("Force d'attaque : "))
    agilite = int(input("Agilité (1-50) : "))
    armure = int(input("Points d'armure : "))
    
    nouveau = {
        'nom': nom, 
        'pv': pv, 
        'force': force, 
        'agilite': agilite, 
        'armure': armure
    }
    equipe.append(nouveau)

# Exemple d'utilisation pour 3 guerriers
for i in range(3):
    creer_guerrier(equipe)
```

**Question 3 :**
Créer la fonction `trier_guerrier(equipe)` qui prend une équipe en paramètre et modifie cette liste pour la trier selon la somme des caractéristiques des guerriers.


```python
def trier_guerriers(equipe):
    """Trie l'équipe par ordre décroissant de score total (Tri par insertion)."""
    n = len(equipe)
    for i in range(1, n):
        cle = equipe[i]
        k = i
        # On compare les scores totaux pour un tri décroissant [cite: 118]
        while k > 0 and calculer_score(equipe[k-1]) < calculer_score(cle):
            equipe[k] = equipe[k-1] # Décalage vers la droite [cite: 120]
            k = k - 1
        equipe[k] = cle

# Test du tri
trier_guerriers(equipe)
print("Équipe triée par puissance :", equipe)
```

```python
trier_guerriers(equipe)
print(equipe)
```

**Question 4 :**

Le combat oppose un attaquant et un défenseur.

Règles logiques à implémenter :

* Esquive : Si l'agilité du défenseur est supérieure à un nombre aléatoire (entre 1 et 50), il esquive le coup.

* Dégâts : Si le coup porte, les dégâts sont égaux à : force de l'attaquant - armure du défenseur.

* Minimum : Les dégâts ne peuvent pas être inférieurs à 0.


```python
import random

def attaquer(attaquant, defenseur):
    """Simule une attaque selon les règles d'esquive et d'armure."""
    print(f"⚔️ {attaquant['nom']} attaque {defenseur['nom']} !")
    
    # Règle d'esquive : agilité vs nombre aléatoire 1-50 
    if defenseur['agilite'] > random.randint(1, 50):
        print(f"💨 Esquivé ! {defenseur['nom']} est trop agile.")
    else:
        # Dégâts = Force - Armure (minimum 0) 
        degats = attaquant['force'] - defenseur['armure']
        if degats < 0:
            degats = 0
            
        defenseur['pv'] -= degats
        print(f"💥 {degats} dégâts infligés ! PV restants de {defenseur['nom']} : {max(0, defenseur['pv'])}")
```

**Question 5 :** Le combat

Simuler un combat au tour par tour entre votre guerrier et un monstre de votre création (ex: un Dragon avec 200 PV) 

```python
def combat(guerrier1, guerrier2):
    """Simule un combat entre deux guerriers jusqu'à la mort de l'un d'eux."""
    print(f"--- ⚔️ DÉBUT DU DUEL : {guerrier1['nom']} vs {guerrier2['nom']} ---")
    
    # La boucle continue tant que les DEUX sont vivants
    while guerrier1['pv'] > 0 and guerrier2['pv'] > 0:
        # On utilise la fonction attaquer de la Q4
        attaquer(guerrier1, guerrier2)
        
        # On vérifie si le deuxième est toujours en vie avant qu'il ne réplique
        if guerrier2['pv'] > 0:
            attaquer(guerrier2, guerrier1)
            
    # Affichage du résultat final
    if guerrier1['pv'] <= 0:
        print(f"💀 {guerrier1['nom']} est tombé au combat !")
    else:
        print(f"🏆 {guerrier2['nom']} est vaincu, {guerrier1['nom']} l'emporte !")
```

**Question 6 :** Combat en équipe

Simuler un combat au tour par tout entre deux équipe de guerriers ou de monstres.
* chaque équipe envoie ses guerriers du plus puissant au moins puissant
* chaque combat est 1 vs 1
* le guerrier qui commence à attaquer est choisi au hasard
* l'attaquant porte un coup, génère des dégats
* si le défenseur est toujours vivant, c'est à son tour de devenir attaquant
* le vainqueur reste et affronte le guerrier suivant de l'équipe adverse.
* le combat s'arrête lorsqu'une équipe n'a plus de combattant.

Vous pouvez utiliser pour vos tests les équipes suivantes:



```python
guerrier1 = {'nom': 'Aethelgard', 'pv': 120, 'force': 10, 'agilite': 5, 'armure': 15}
guerrier2 = {'nom': 'Slythe', 'pv': 70, 'force': 25, 'agilite': 45, 'armure': 2}
guerrier3 = {'nom': 'Valeria', 'pv': 100, 'force': 18, 'agilite': 20, 'armure': 10}
guerrier4 = {'nom': 'Grognar', 'pv': 150, 'force': 22, 'agilite': 10, 'armure': 5}
equipe = [guerrier1, guerrier2, guerrier3, guerrier4]

monstre1 = {"nom": "Gobelin", "pv": 50, "force": 8, "agilite": 12, "armure": 3}
monstre2 = {"nom": "Orc", "pv": 80, "force": 15, "agilite": 8, "armure": 5}
monstre3 = {"nom": "Troll", "pv": 120, "force": 20, "agilite": 5, "armure": 10}
monstres = [monstre1, monstre2, monstre3]

```

```python
def combat_equipe(equipe1, equipe2):
    """Simule un combat entre deux équipes de guerriers et de monstres."""
    
    # 1. On trie les deux équipes par puissance décroissante (Tri Insertion AP08)
    trier_guerriers(equipe1)
    trier_guerriers(equipe2)
    
    i = 0 # Index pour l'équipe 1
    j = 0 # Index pour l'équipe 2
    
    # La bataille continue tant qu'il reste des combattants dans les DEUX équipes
    while i < len(equipe1) and j < len(equipe2):
        g1 = equipe1[i]
        g2 = equipe2[j]
        
        print(f"\n>>> PROCHAIN DUEL : {g1['nom']} vs {g2['nom']}")
        
        # On lance le duel (on peut copier la logique de la Q5 ici)
        # On choisit aléatoirement qui porte le premier coup du duel
        if random.choice([True, False]):
            attaquant_actuel, defenseur_actuel = g1, g2
        else:
            attaquant_actuel, defenseur_actuel = g2, g1
            
        while g1['pv'] > 0 and g2['pv'] > 0:
            attaquer(attaquant_actuel, defenseur_actuel)
            # On inverse les rôles pour le prochain coup
            attaquant_actuel, defenseur_actuel = defenseur_actuel, attaquant_actuel
            
        # 2. Gestion des survivants : le perdant est éliminé, le vainqueur reste
        if g1['pv'] <= 0:
            print(f"❌ {g1['nom']} est éliminé.")
            i += 1 # On passe au guerrier suivant de l'équipe 1
        else:
            print(f"❌ {g2['nom']} est éliminé.")
            j += 1 # On passe au monstre suivant de l'équipe 2

    # 3. Annonce de l'équipe victorieuse
    if i < len(equipe1):
        print("\n🏁 VICTOIRE FINALE : L'équipe des Guerriers a survécu !")
    else:
        print("\n🏁 DÉFAITE : L'équipe des Monstres a tout ravagé...")
```
