## 1. Anatomie d'une fonction

Pour définir une fonction, on utilise le mot-clé `def` suivi du nom de la fonction et de parenthèses.

```python
def nom_de_la_fonction(parametre1, parametre2):
    # Bloc d'instructions (indentation obligatoire)
    resultat = parametre1 + parametre2
    return resultat
```


Prenons un exemple avec la fonction `somme` qui prend deux paramètres et renvoie la somme des valeurs passées en paramètre. Il est enseuite possible de récupérer ce qui est renvoyé par cette fonction.

Ici:
* le **nom** de la fonction est `somme`
* `a` et `b` sont les paramètres de cette fonction.
* `y`, `a` et `b` sont des variables locales.
* `v1` est une variable globale.


Lors de l'appel `v1 = somme(6, 7)`:
* 6 et 7 sont des arguments
* on appelle la fonction `somme(6, 7)`
* la valeur renvoyée par la fonction est stockée dans `v1`.

````python
def somme(a, b):
    ''' Renvoie la somme de a et b '''
    y = a + b
    return y

v1 = somme(6, 7)
v2 = somme(-4, 9)
print(v1)
print(v2)
print(v1 + v2)
````

## 2. Tester la valeur d'une variable


Il est parfois compliqué de vérifier si la fonction écrite donne le bon résultat pour toutes les combinaisons possibles de paramètres. On va donc la tester sur des exemples particuliers et dire que si ces cas particuliers ont été assez bien choisis, la fonction est correcte.

On utilise la méthode des `assert`. Elle s'écrit:

```python
assert condition , "Message d'erreur"
```

Si la condition écrite est vraie (`True`), il ne se passe rien. Si elle est fausse (`False`), le `Message d'erreur` est affiché et le programme est immédiatement interrompu.

Tester les trois cellules suivantes



````python
a = 8
assert ajouter(6, 7) == 13                          # Message d'erreur affiché si ajouter(6, 7) n'est pas égal à 13
assert type(a) == int, "a doit être un entier"      # Message d'erreur affiché si a n'est pas de type int
assert a > 0, "a doit être strictement positif"     # Message d'erreur affiché si a n'est pas strictement positif
````
