### **Introduction : La boucle FOR**

La boucle FOR est une **boucle bornée**. On sait avant de commencer combien de boucle nous allons effectuer.

#### Itération avec un itérateur `i`:

```python
for i in range(0, 7):
    # i prend toutes les valeurs entre 0 et 6
    # il y aura 7 boucles

for i in range(3, 8):
    # i prend toutes les valeurs entre 3 et 7
    # il y aura 5 boucles

for i in range(4):
    # i prend toutes les valeurs entre 0 et 4
    # il y aura 4 boucles

```

#### Itération sur un itérable avec les indices:

```python
mot = 'abracadabra'
n = len(mot)    # n vaut 11

for i in range(n):
    # i prend toutes les valeurs entre 0 et n-1
    # il y aura n boucles
    print(mot[i])   # Affichage de la lettre d'indice i
```

#### Itération sur un itérable avec les éléments:

```python
mot = 'abracadabra'

for lettre in mot:
    # lettre prend pour valeur chaque lettre de mot
    # il y aura 11 boucles
    print(element)   # Affichage de la lettre element
```

### **Introduction : Les chaines de caractères**

#### Indexation des chaines de caractères

Chaque caractère a un indice commençant à 0

```python 
mot = 'abracadabra'
n = len(mot)    # n vaut 11

# mot[0] vaut 'a'
# mot[4] vaut 'c'
# mot[n-1] vaut 'a'
# mot[n] n'existe pas
```

#### Concaténation des chaines de caractères

```python
mot1 = 'Hello'
mot2 = ' World'
mot3 = mot1 + mot2
# mot3 vaut 'Hello World'
```
