## Exercice 1 : L'approche Naïve
L'approche récursive pure sans stockage.

```python
def rugby_naif(n):
    if n == 0: return 1
    if n < 0: return 0
    # On additionne les possibilités pour chaque type de point
    return rugby_naif(n-3) + rugby_naif(n-5) + rugby_naif(n-7)

print(f'Manières pour 15 pts : {rugby_naif(15)}')
```

## Exercice 2 : Top-Down (Mémoïsation)
On utilise un dictionnaire pour éviter de recalculer les scores déjà rencontrés.

```python
def rugby_memo(n, memo=None):
    if memo is None: memo = {}
    if n == 0: return 1
    if n < 0: return 0
    
    # 1. Vérification du cache
    if n in memo:
        return memo[n]
    
    # 2. Calcul et stockage
    memo[n] = rugby_memo(n-3, memo) + rugby_memo(n-5, memo) + rugby_memo(n-7, memo)
    return memo[n]

print(f'Manières pour 50 pts : {rugby_memo(50)}')
```

## Exercice 3 : Bottom-Up (Tabulation)
On remplit un tableau de manière itérative.

```python
def rugby_tabulation(n):
    dp = [0] * (n + 1)
    dp[0] = 1
    points = [3, 5, 7]
    
    for i in range(1, n + 1):
        for p in points:
            if i >= p:
                dp[i] += dp[i - p]
                
    return dp[n]

print(f'Manières pour 100 pts : {rugby_tabulation(100)}')
```

## Exercice 4 : Combinaisons Uniques
En inversant les boucles, on s'assure de ne pas compter les permutations (ex: 3+5 et 5+3 deviennent une seule combinaison {3,5}).

```python
def rugby_uniques(n):
    dp = [0] * (n + 1)
    dp[0] = 1
    points = [3, 5, 7]
    
    for p in points: # On traite un type de point à la fois
        for i in range(p, n + 1):
            dp[i] += dp[i - p]
    return dp[n]

print(f'Combinaisons uniques pour 8 pts : {rugby_uniques(8)} (Attendu: 1)')
```

## Exercice 5 & 6 : Benchmark et Graphique
Comparaison visuelle des performances.

```python
import time
import matplotlib.pyplot as plt

scores = range(1, 36)
temps_naif = []
temps_memo = []
temps_dp = []

for n in scores:
    # Naïf
    t0 = time.time()
    rugby_naif(n)
    temps_naif.append(time.time() - t0)
    
    # Memo
    t0 = time.time()
    rugby_memo(n, {})
    temps_memo.append(time.time() - t0)
    
    # Tabulation
    t0 = time.time()
    rugby_tabulation(n)
    temps_dp.append(time.time() - t0)

plt.figure(figsize=(10, 6))
plt.plot(scores, temps_naif, label='Naïf', color='red')
plt.plot(scores, temps_memo, label='Mémoïsation', color='green')
plt.plot(scores, temps_dp, label='Tabulation', color='blue')
plt.title('Explosion de la complexité exponentielle')
plt.xlabel('Score')
plt.ylabel('Temps (s)')
plt.legend()
plt.grid(True)
plt.show()
```
