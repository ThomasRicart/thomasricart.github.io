```python
import time

def benchmark_rugby(n):
    print(f"--- Benchmark pour n = {n} ---")
    
    # 1. Mesure du Naïf
    start = time.time()
    res_naif = rugby_naif(n)
    end = time.time()
    print(f"Naïf       : {res_naif} manières | Temps : {end - start:.5f} secondes")
    
    # 2. Mesure de la Mémoïsation
    start = time.time()
    res_memo = rugby_memo(n, memo={})
    end = time.time()
    print(f"Mémoïsation : {res_memo} manières | Temps : {end - start:.5f} secondes")
    
    # 3. Mesure de la Tabulation
    start = time.time()
    res_tab = rugby_tabulation(n)
    end = time.time()
    print(f"Tabulation  : {res_tab} manières | Temps : {end - start:.5f} secondes")

# Testez avec n = 35
benchmark_rugby(35)
```
